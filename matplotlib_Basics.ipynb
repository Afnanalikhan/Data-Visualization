# Line Graph with Subplots

some time we want to display two lines side by side rather than in the same set of x-axis and y-axis 
we have multiple axis we call each set of axes a subplot.


```python
# Subplot
# The command plt.subplot() needs three arguments to be passed into it
# ) The number of rows of subplot
#) The number of Column of subplot
# ) The index of the subplot we want to create

# For instance, the cammand plt.subplot(2,3,20 would create 'subplot2' 
```

```python
#Data sets
x = [1,2,3,4]
y = [1,2,3,4]

#First subplot
plt.subplot(1,2,1)
plt.plot(x,y, color='green', linestyle='--')
plt.title('First Subplot')

#Second subplot
plt.subplot(1,2,2)
plt.plot(x,y, color='steelblue', marker='s')
plt.title('Second Subplot')

plt.show()
```

```python
#Task
#We have define the list months, temperature, flight_to_hawai
# using the subplot command, plot temperature VS months in the left box of a figure that has 1 row wit 2 columns
months = range(12)
temperature = [36,36,39,52,61,72,77,75,68,57,48,48]
flight_to_hawai = [1200,1300,1100,1450,850,750,400,450,400,860,990,1000]

plt.subplot(1,2,1)
plt.plot(temperature,months, color='orange')
plt.subplots_adjust(wspace=0.35)
plt.xlabel('Temperature')
plt.ylabel('Month')
plt.title('Month and Temp demonstration')

plt.subplot(1,2,2)
plt.plot(flight_to_hawai,temperature, color='brown')
plt.subplots_adjust(wspace=0.35)
plt.xlabel('Flight_to_hawai')
plt.ylabel('Temperature')
plt.title('Fligh and Temp demonstration')

plt.show()
```
